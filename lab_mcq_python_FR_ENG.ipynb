# **Question 2: In Python, how do you create a list named `myList` containing the numbers 1, 2, and 3?**


**Answer A: myList = [1, 2, 3]**

**ENG**: This is the correct way to define a list in Python. Lists are defined using square brackets.

**FR**: C’est la bonne manière de définir une liste en Python. Les listes sont créées avec des crochets.


````python
myList = [1, 2, 3]
print(type(myList), myList)
````

**Answer B: myList <- c(1, 2, 3)**

**ENG**: This is R syntax, not Python. The assignment operator `<-` and the function `c()` are not valid in Python.

**FR**: Ceci est une syntaxe du langage R, pas de Python. L’opérateur `<-` et la fonction `c()` ne sont pas valides en Python.


````python
myList <- c(1, 2, 3)
````

**Answer C: myList = (1, 2, 3)**

**ENG**: This creates a tuple, not a list. A tuple is immutable, whereas a list is mutable.

**FR**: Ceci crée un tuple, pas une liste. Un tuple est immuable, tandis qu’une liste est mutable.


````python
myList = (1, 2, 3)
print(type(myList), myList)
````

**Answer D: myList = {1, 2, 3}**

**ENG**: This creates a set, not a list. Sets are unordered and do not allow duplicate values.

**FR**: Ceci crée un ensemble (set), pas une liste. Les ensembles sont non ordonnés et ne permettent pas de doublons.


````python
myList = {1, 2, 3}
print(type(myList), myList)
````

**Summary (ENG)**:

**A**: Correct – Valid Python list  
**B**: Invalid – R syntax  
**C**: Invalid – Tuple, not a list  
**D**: Invalid – Set, not a list

**Résumé (FR)**:

**A** : Correct – Liste Python valide  
**B** : Invalide – Syntaxe R  
**C** : Invalide – Tuple, pas une liste  
**D** : Invalide – Ensemble (set), pas une liste
